```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm


np.random.seed(42)

# generate data like sin curve
def generate_data(n, sigma):
    x = np.random.uniform(-1, 1, n)
    y = np.sin(np.pi * x) + np.random.normal(0, sigma, n)
    return x, y

train_data = generate_data(20, 0.3)
test_data = generate_data(1000, 0.3)
plt.scatter(train_data[0], train_data[1], c='tab:blue', label='train data')
plt.scatter(test_data[0], test_data[1], marker=".", c='tab:orange', label='test data', alpha=0.3)

xx = np.linspace(-1, 1, 100)
yy = np.sin(np.pi * xx)
plt.plot(xx, yy, c='tab:green', label='sin curve')
plt.legend()
plt.show()
```

```python
# 
model = svm.SVR(kernel='poly', C=100, epsilon=1e-3, gamma=1, degree=21, coef0=1.0)
model.fit(train_data[0].reshape(-1, 1), train_data[1])

x_reg = np.linspace(-1, 1, 100)
y_reg = model.predict(x_reg.reshape(-1, 1))

plt.scatter(train_data[0], train_data[1], c='tab:blue', label='train data')
plt.plot(xx, yy, c='tab:green', label='sin curve')
plt.plot(x_reg, y_reg, label='Regression curve', color='red')
plt.ylim(-2, 2)
plt.legend()
```
